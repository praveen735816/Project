```python
x=df.iloc[:,:-1]
x.head(2)
y=df.iloc[:,-1]
```

```python
y.value_counts()
```

```python
def calc_vif(x):
    vif=pd.DataFrame()
    vif['variables']=x.columns
    vif['vif factor']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return(vif)
```

```python
#now we have to balance the data set
# for that import SMOTE
from imblearn.over_sampling import SMOTE
x, y = SMOTE().fit_resample(x, y)
```

```python
y.value_counts()
```

```python
# its time to feed the data to ml model
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.4, random_state=1)
```

```python
# now we can feed the data to moodels
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, roc_auc_score
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
lr=LogisticRegression()
rf,ad,dt=RandomForestClassifier(), AdaBoostClassifier(), DecisionTreeClassifier()
```

```python

ml=[lr,rf,ad,dt]
for i in ml:
    i.fit(xtrain,ytrain)
    test_pre=i.predict(xtest)
    print("Below values of",i)
    print("test accuracy_score of",accuracy_score(ytest,test_pre))
    print("cross validation score", cross_val_score(i,x,y,cv=15).mean()*100)
    print("confusion_matrix of",'\n', confusion_matrix(ytest,test_pre))
    print("classification_report of",'\n',classification_report(ytest,test_pre))
```

```python
# RandomForestClassifier performs best with 66.3% accuracy, 70% cv score
```

```python
#now we will select the best random state value where the ml model having the exact fit
```

```python
for i in range(1,100):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3, random_state=i)
    rf.fit(xtrain,ytrain)
    train_pre=rf.predict(xtrain)
    test_pre=rf.predict(xtest)
    if accuracy_score(ytest,test_pre)==accuracy_score(ytrain,train_pre):
        print("Random state of",i,"the model having exact fit")
        print("test accuracy_score of",accuracy_score(ytest,test_pre))
        print("cross validation score", cross_val_score(rf,x,y,cv=15).mean()*100)
        print("classification_report of",'\n',classification_report(ytest,test_pre))

    
```
