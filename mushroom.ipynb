```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import sklearn
```

```python
print(dir(sklearn.preprocessing))
```

```python
#importing the data set into jupyter
m=pd.read_csv("mushroom.csv")
m.head(5)
```

```python
one=OneHotEncoder()
lb=LabelEncoder()
```

```python
#converting the data into numeric form
m['class']=lb.fit_transform(m['class'])
m['class']
```

```python
#converting the comlete data into numeric form by using the lable encoder
for col in  m:
    m[col]=lb.fit_transform(m[col])
m
    
```

```python
m.describe(include='all')
```

```python
#separating the data based on the dependent vaiables as y and independent variables as x
x=m.iloc[:,1:]
y=m.iloc[:,0]
```

```python
x.shape,y.shape
```

```python
y
```

```python
x.size,y.size
```

```python
x.ndim, y.ndim
```

```python
#spliting the data for training and testing
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=45)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
lg=LogisticRegression()
```

```python
#since the out put is e or p we using the logistic regression
#fitting the data into the equation for training
lg.fit(xtrain,ytrain)
# x variable for test
pre=lg.predict(xtest)
#out put
pre
```

```python
ytest=np.array(ytest)
ytest=ytest.reshape(1,-1)
ytest
```

```python
#checking the model accuracy
accuracy_score(ytest,pre)
```

```python
#checkingthe true positive and true negative values
confusion_matrix(ytest,pre)
```

```python
#checking the all the other values in the logisticsregression
print(classification_report(ytest,pre))
```
